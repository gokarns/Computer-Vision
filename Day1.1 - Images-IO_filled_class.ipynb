Let's now load our first image

```python
# We don't need to do this again, but it's a good habit
import cv2 

# Load an image using 'imread' specifying the path to image
# You can either point to a relative path or else a full absolute path
input_img = cv2.imread('./images/input.jpg')

# Our file 'input.jpg' is now loaded and stored in python 
# as a varaible we named 'image'

# To display our image variable, we use 'imshow'
# The first parameter will be title shown on image window
# The second parameter is the image varialbe
# You can create as many windows as you wish, but with different window names.
cv2.imshow('Our sample image', input_img)

# 'waitKey' allows us to input information when a image window is open
# By leaving it blank it just waits for anykey to be pressed before 
# continuing. By placing numbers (except 0), we can specify a delay for
# how long you keep the window open (time is in milliseconds here)
cv2.waitKey(5000)

# This closes all open windows 
# Failure to place this will cause your program to hang
# If you want to destroy any specific window, use the function cv2.destroyWindow() 
# where you pass the exact window name as the argument.
cv2.destroyAllWindows()
```

```python
# Same as above without the comments

import cv2 
input_img_beatle = cv2.imread('./images/beatle.jpg')
cv2.imshow('Our sample image beatle', input_img_beatle)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
# A case where you can already create a window and load image to it later

import cv2 
#read an image
beatle_img = cv2.imread('./images/beatle.jpg')

# You can specify whether window is resizable or not
# named window :- resize the windows, track the windows based a name/id
# cv2.WINDOW_NORMAL: window is resizable manually
cv2.namedWindow('first_image_window', cv2.WINDOW_NORMAL)

# Display and assign images to a window 
cv2.imshow('first_image_window', beatle_img)

cv2.waitKey(5000)
cv2.destroyAllWindows()
```

### Import and display images using matplotlib

```python
# this statemnet is required to show plots in line in the notebook
%matplotlib inline

import numpy as np
import cv2
# Import pyplot package from matplotlib
from matplotlib import pyplot as plt

#Load the image file
img = cv2.imread('./images/beatle.jpg')
#show the image using plt.imshow
plt.imshow(img)
plt.show()

```

```python
# to hide tick values on X and Y axis
plt.imshow(img)
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
img.shape
```

```python
# Image is discolored when plotted using matplot lin plt function
# the issue is with the bgr->rgb color transformation

# read the input image into its constituents
b,g,r = cv2.split(img)

# transfor the image into rgb
img_transformed = cv2.merge([r,g,b])

plt.imshow(img_transformed)
plt.xticks([]),plt.yticks([])
plt.show()
```
