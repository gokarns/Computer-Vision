```python
#create a function getSketch()
#Input:- inp_BGR image
#Output :- sketched_image
#Sketching involves four basic steps
%matplotlib inline

import cv2
import numpy as np
import ImageUtils


img = cv2.imread('D:/farzad/Artificial Intelligence/CV/fromTutor-20180901T044023Z-001/fromTutor/images/obama.jpg')
ImageUtils.show_img_plt(img)


#Step1
#Convert BGR to Grayscale
#cv2.COLOR_BGR2GRAY

gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ImageUtils.plot_grayscale_imshow(gray_image, 5)

#Step2
#Clean image using Gaussian blur
#cv2.GaussianBlur

GB = cv2.GaussianBlur(gray_image,(5,5),0)
ImageUtils.plot_grayscale_imshow(GB, 5)

#Step3
#Extract edges using Canny's Edge detection
#cv2.Canny

canny_edge= cv2.Canny(img, 50,120)
ImageUtils.plot_grayscale_imshow(canny_edge,5)

#Step4
#Invert binarized image
#cv2.THRESH_BINARY_INV
adaptive_threshold = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 3,5)
ImageUtils.plot_grayscale_imshow(adaptive_threshold, 5)
```

```python
#Read an input image
#Diplay the image here

img2 = cv2.imread('D:/farzad/Artificial Intelligence/CV/fromTutor-20180901T044023Z-001/fromTutor/images/obama.jpg')
ImageUtils.show_img_plt(img2)


#call getSketch()
#display sketched image here
sb,sc = cv2.pencilSketch(img2, sigma_s=60, sigma_r=0.07, shade_factor=0.05)
ImageUtils.plot_grayscale_imshow(sb,5)
ImageUtils.show_img_plt(sc)
```
