#### Affine transform just shifts the position of an image
#### cv2.warpAffine

```python
%matplotlib inline
import cv2
import numpy as np
import image_utils
```

```python
inp_image = cv2.imread('images/input.jpg')
image_utils.show_img_plt(inp_image)
```

```python
height, width = inp_image.shape[:2]
translate_x = 100
translate_y = 200


#       | 1 0 Tx |
#  T  = | 0 1 Ty |

T = np.float32([[1,0,translate_x],[0,1,translate_y]])

#warpAffine

translated_img = cv2.warpAffine(inp_image, T, (width,height))

image_utils.show_img_plt(translated_img)
```

#### Rotations

```python
# M = [[cos(theta), -sin(theta)],[sin(theta),cos(theta)]]

#rotate the image by 90 w.r.t center
# no scaling

M = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)

rotated_img = cv2.warpAffine(inp_image, M, (width,height))

image_utils.show_img_plt(rotated_img)
```

#### ANother method for rotation
#### Removes the black padded pixels

```python
rotated_img_transpose = cv2.transpose(inp_image)
image_utils.show_img_plt(rotated_img_transpose)
```

#### Horizontal flip
#### 180

```python
M = cv2.getRotationMatrix2D((width/2, height/2), 180, 1)

rotated_img = cv2.warpAffine(inp_image, M, (width,height))

image_utils.show_img_plt(rotated_img)
```

#### Horizontal flip using the flip function

```python
horizontal_flip = cv2.flip(inp_image,1)
image_utils.show_img_plt(inp_image)

image_utils.show_img_plt(horizontal_flip)
```

#### Scaling is done using the function, resize

```python
#resize (inp_image, <exact pixel size>, fx, fy )

scaled_image1 = cv2.resize(inp_image, None, fx=0.75, fy=0.75)
image_utils.show_img_plt(scaled_image1)
```

```python
scaled_image2 = cv2.resize(inp_image, None, fx=2.5, fy=2.5, interpolation = cv2.INTER_CUBIC)
image_utils.show_img_plt(scaled_image2)
```

```python
scaled_image3 = cv2.resize(inp_image, (100,50), interpolation = cv2.INTER_AREA)
image_utils.show_img_plt(scaled_image3)
```
