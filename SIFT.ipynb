```python
%matplotlib inline

import cv2
import numpy as np
import ImageUtils


img = cv2.imread('D:/farzad/Artificial Intelligence/CV/fromTutor-20180901T044023Z-001/fromTutor/images/input.jpg')
ImageUtils.show_img_plt(img)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ImageUtils.plot_grayscale_imshow(gray_img,5)
```

```python
sift_detector = cv2.xfeatures2d.SIFT_create()

keypoints = sift_detector.detect(gray_img,None)
len(keypoints)
```

```python


image_out = cv2.drawKeypoints(img,keypoints,outImage=np.array([]),color=(0,0,255),flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.imshow('Output img',image_out)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
surf_detector = cv2.xfeatures2d.SURF_create(7000)
```

```python
#surf_detector.hessianThreshold = 500
keypoints, desc = surf_detector.detectAndCompute(gray_img, None)
len(keypoints)

image_out = cv2.drawKeypoints(img,keypoints,outImage=np.array([]),color=(0,0,255),flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.imshow('Output img',image_out)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
